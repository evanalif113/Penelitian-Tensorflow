```python
import pandas as pd
import io
import requests

# --- KONFIGURASI ---
API_KEY = '5CNVUZFKADUX7W7V5S533MYBG' # Jangan pakai yang bocor tadi
LAT = -7.7364686311041
LON = 109.64606580850659
START_DATE = '2025-01-01'
END_DATE = '2025-10-31'

# URL Builder (Pakai METRIC!)
base_url = "https://weather.visualcrossing.com/VisualCrossingWebServices/rest/services/timeline"
query_url = f"{base_url}/{LAT},{LON}/{START_DATE}/{END_DATE}?unitGroup=metric&include=days&key={API_KEY}&contentType=csv"

print("Sedang download data dari Visual Crossing...")

# Request Data
response = requests.get(query_url)

if response.status_code == 200:
    # Baca CSV dari text response langsung ke Pandas
    df_vc = pd.read_csv(io.StringIO(response.text))
    
    # Ubah kolom tanggal jadi datetime
    df_vc['datetime'] = pd.to_datetime(df_vc['datetime'])
    df_vc = df_vc.set_index('datetime')
    
    print("✅ Berhasil!")
    print(df_vc[['temp', 'precip', 'humidity', 'sealevelpressure']].head())
    
    # Penjelasan Kolom Visual Crossing (Metric):
    # temp: Suhu Rata-rata (°C)
    # precip: Curah Hujan (mm)
    # humidity: Kelembapan (%)
    # sealevelpressure: Tekanan QNH (hPa)
    
else:
    print(f"❌ Gagal: {response.status_code}")
    print(response.text)
```
